```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure()
x1 = [2, 4, 30, 45, 60]
x2 = [45, 44, 36, 22, 1]
plt.plot(x1, label='Anakin Darkside Meter', color='red')
plt.plot(x2, label='Younglings in Temple', color='blue')
plt.legend()
sns.set()
# sns.set(style='whitegrid')
sns.set_style('darkgrid')
sns.set(font='serif')
plt.show()
```

```python
flight = sns.load_dataset('flights')
flight
```

```python
sns.get_dataset_names()
```

```python
sns.relplot(x='year', y='passengers', data=flight)
```

```python
year_sums = flight.groupby('year', as_index=False).sum()
```

```python
sns.relplot(x='year', y='passengers', data=year_sums)
```

```python
sns.barplot(x='year', y='passengers', data=flight)
```

```python
g = sns.barplot(x='month', y='passengers', data=flight, ci=None)
g.set_xticklabels(labels=flight['month'], rotation=45)
```
